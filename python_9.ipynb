```
Problem:

Write a Python function that:

- Adds the last element of list data to the beginning of tuple_data and returns the result as a new tuple.

- Removes the number 70 from set_data if it exists and returns the updated set.

- Combines all the data (from data, tuple_data, and set_data) into a single list and removes any duplicate values.

إضافة آخر عنصر من قائمة data إلى بداية الـ tuple المسمّى tuple_data وإرجاع النتيجة كـ tuple جديد.

إزالة الرقم 70 من المجموعة (set) المسماة set_data إذا كان موجودًا، ثم إرجاع الـ set بعد التحديث.

دمج جميع البيانات من (data, tuple_data, set_data) في قائمة واحدة (list) مع إزالة القيم المكررة.
```


````python

def manipulate_data(data,tuple_data,set_data):
  new_tuple=(data[-1],)+tuple_data
  set_data.discard (70)
  combined_list=list(data)+list(new_tuple)+list(set_data)
  combined_list=list(set(combined_list))
  return new_tuple,set_data,combined_list
data = [10,20,30,40,50,50]
tuple_data = (5, 15, 25, 35)
set_data = {60, 70, 80}

result=manipulate_data(data,tuple_data,set_data)
print(result)
````

If n is odd, prints the first n odd numbers.

````python
def test_number():
  number=int(input('please enter the number to be tested:'))
  if number <1:
    print('the number must be larger than 1')
  else:
    if number %2==0:
      for i in range(0,2*number,2):
        print(i)
    else:
      for i in range(1,2*number,2):
        print(i)
test_number()
````

Problem 3:

Write a Python function that: Accepts a list of integers, Returns a new list that contains:

The sum of all even numbers.

The product of all odd numbers.

The total number of positive numbers.

المسألة:

اكتب دالة بلغة بايثون تقوم بما يلي: تستقبل قائمة من الأعداد الصحيحة، وتُرجع قائمة جديدة تحتوي على:

مجموع جميع الأعداد الزوجية.

حاصل ضرب جميع الأعداد الفردية.

العدد الإجمالي للأعداد الموجبة.



````python
numbers = [-2, 0, 4, 1, 5, -5]
def test_list():
  sum_even = 0
  prod_odd = 1
  count_positive = 0
  for num in numbers:
    if num % 2 == 0:
      sum_even += num
    else:
      prod_odd *= num
    if num > 0:
      count_positive += 1

  print('Sum of all even numbers: ',sum_even)
  print('Product of all odd numbers: ',prod_odd)
  print('Count of all positive numbers: ', count_positive)
test_list()
````
