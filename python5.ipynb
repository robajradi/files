```python
def prod_function (num1,num2): # عشان اعرف دالة لازم اكتب دف واسم الدالة يفضل اسم الدالة يكون معبر مثلا بحط دالة تضرب رقمين في بعض عشان كذا كتبنا برود فونكشن الدالة هذي بتقبل مني رقمين واسمي الرقمين اول
  result=num1*num2
  return result #هذا الاوت بوت حق الدالة
prod_function (3,5) #هنا عملت الاستدعاء حق الدالة احط اسم الدالة والانبوت
```
