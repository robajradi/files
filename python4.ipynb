Student=[
{'name':'Alice','grades':[85,78,92]},
{'name':'Bob','grades':[58,60,45]},
{'name':'Charlie','grades':[90,91,89]},
for each student
1-calculate the average grade,
2-add a key 'status':
'pass' the average is > 60,
'faill' otherwise.
3-keep a set of students who scored 90 or above in any subject
4-print the name and status of each student
5-print the set of high performers


```python
#حبه حبه بنحل السؤال اول شي نطلع المتوسط وعشان نطلع المتوسط لازم نفرق اللسته عشان نرتبها
students=[ {'name':'Alice','grades':[85,78,92]}, {'name':'Bob','grades':[58,60,45]}, {'name':'Charlie','grades':[90,91,89]}]
#first=students[0]
#first['grades']
#students[0]['grades']
#sum(students[0]['grades'])/len(students[0]['grades'])
for stud in students:#هنا اقوله لكل طالب داخل قائمة الطلبة
  print(sum(stud['grades'])/len(stud['grades'])) #هنا اقوله الامر ايش يسوي
```

```python
#الحين نحل ثاني سؤال
# 2-add a key 'status': 'pass' the average is > 60, 'faill' otherwise.
students=[ {'name':'Alice','grades':[85,78,92]}, {'name':'Bob','grades':[58,60,45]}, {'name':'Charlie','grades':[90,91,89]}]
for stud in students:#هنا اقوله لكل طالب داخل قائمة الطلبة
  average_score=(sum(stud['grades'])/len(stud['grades'])) #هنا اقوله الامر ايش يسوي
  print(average_score)
  if average_score>=60: #هنا حطيت الاف زي الي يقول اذا كان مستوى الطالب اكبر من كذا وتحته يكون الجواب
    stud['status']='pass'
  else:
    stud['status']='faill'
students
```

```python
#الحين نحل ثالث سؤال
# 3-keep a set of students who scored 90 or above in any subject
students=[ {'name':'Alice','grades':[85,78,92]}, {'name':'Bob','grades':[58,60,45]}, {'name':'Charlie','grades':[90,91,89]}]
set_stud=set()# هنا لازم اعرف السيت الي تحت عشان تمشي وليه حطينا set عشان لاتسمح بالتكرار
#مثلا تشارلي عنده درجتين 90 وبيطبع اثنين لو حطينا لست بس لان سيت لا تقبل بالتكرار حطيناها عشان ماتطبع اسمه مرتين
for stud in students:#هنا اقوله لكل طالب داخل قائمة الطلبة
  average_score=(sum(stud['grades'])/len(stud['grades'])) #هنا اقوله الامر ايش يسوي
  print(average_score)
  if average_score>=60: #هنا حطيت الاف زي الي يقول اذا كان مستوى الطالب اكبر من كذا وتحته يكون الجواب
    stud['status']='pass'
  else:
    stud['status']='faill'
print(students)
for stud in students: #اول فور لتعريف الطلبة
  for grad in stud['grades']:# ثاني فور لتعريف الدرجات لكل طالب
    if grad>=90: #هنا اقوله اذا كانت الدرجات اكبر من او يساوي 90
      set_stud.add(stud['name']) # هنا احط له امر انك لازم تطبع اسماء الي ماخذين درجات اكبر من 90
print(set_stud)
for stud in students: #هنا اجابة السؤال الرابع
  print('name:',stud['name'])
  print('status:',stud['status'])
  print('-'*100)
print('set of student with high performers:',set_stud) #أخر سؤال
```
