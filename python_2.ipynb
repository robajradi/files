question

accept a grade from the user and we want to test it
90 -> A+
75, <90 -> A
50, <75 -> B
<50> -> Fail


```python
number=float(input('please enter the number to be tested'))
if number>=90:
  print('the grade is A+')
elif number>=75:
  print('the grade is A')
elif number>=50:
  print('the grade is B')
else:
  print('the grade is fail')
```
