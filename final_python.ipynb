Problem Statement:

Create a function that processes a list of inventory transactions and returns the final inventory status for each item.

Each transaction is a dictionary with:

"item": the name of the item (string)

"type": "add" or "remove"

"quantity": an integer

Your function should:

Track how many of each item are left.

Ensure that the item quantity never goes below 0.

If an item is removed more times than it's added, it should log a warning.

Return a dictionary with:

item name as key

dictionary with:

'stock': current quantity

'warning': True or False

```python
transactions = [
{"item": "apple", "type": "add", "quantity": 10},
{"item": "banana", "type": "add", "quantity": 5},
{"item": "apple", "type": "remove", "quantity": 7},
{"item": "banana", "type": "remove", "quantity": 8},
{"item": "apple", "type": "remove", "quantity": 5},
]
def analyze_products(transactions):
  output_dict = {}

  for trx in transactions:
    item = trx['item']
    trx_type = trx['type']
    qnt = trx['quantity']

    if item not in output_dict:
      output_dict[item] = { 'stock' : 0,'warning': False}

    if trx_type == 'add':
      output_dict[item] ['stock'] += qnt

    elif trx_type == 'remove':
      if output_dict[item] ['stock'] >= qnt:
        output_dict[item] ['stock'] -= qnt

      else:
        output_dict[item] ['stock'] -= qnt

        output_dict[item] ['warning'] = True

  print(output_dict)
analyze_products(transactions)
```
