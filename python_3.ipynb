```python
# initialize
number=1 # اول خطوة الي هي البداية ابغاه يبدء ب 1 بس لاني ابغاه من 1 الى 5
# while condition
while number<=5:
  # code
  print(number)
  # update
  number+=1 # هنا التقفيلة انه ابغاه يحط لي من 1 لين خمسة
```

```python
# الاجابة هذي صحيحة لكن وراها اخطاء مره كثيرة اذا حطيت اكبر من 100 او ارقام سالبةبيقبل معايا وهو المفروض مايقبل
count=1
while count<=5:
  g=float(input("enter your gread"))
  if g>=90:
    print("A+")
  elif g>90 and g>=75:
    print("A")
  elif g>75 and g>=50:
    print("B")
  else:
    print("fail")
  count+=1
```
